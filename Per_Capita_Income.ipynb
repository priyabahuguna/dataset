```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
df = pd.read_csv('https://raw.githubusercontent.com/priyabahuguna/dataset/main/canada_per_capita_income%20-%20Copy.csv')
```

```python
df.head(10)
```

```python
x = df[['year']] #year is independent variable
```

```python
x.head(10)
```

```python
y = df['per capita income (US$)']
```

```python
y.head(10)
```

```python
#Renamed the name of column
df.rename(columns={'per capita income (US$)': 'per capita income'}, inplace=True)
```

# New Section

Predicting Per capita Income in year 2020


```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
year = int(input("Enter the year to predict the price for = "))
```

```python
prediction = model.predict([[year]])

# Print the prediction
print("Prediction:", prediction)
```
